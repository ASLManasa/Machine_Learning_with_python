```python
import random
import sys
```

```python
def checkClash(player, computer, lives, point):                     #Function for checking player's move and computer's move and updating sore and lives
    if(player == 4):   
        sys.exit("You exited from game.")
    else:
        print(steps[player-1], " : ", computer)
        if((computer == "rock" and player == 2) or (computer == "paper" and player == 3) or (computer == "scissor" and player == 1)):
            point += 1
        if((computer == "rock" and player == 3) or (computer == "paper" and player == 1) or (computer == "scissor" and player == 2)):
            lives -= 1
        print("Your point are : ", point, "\nYour Chances remaining are : ",lives)
        return lives, point
    
print("This is Rock, Paper & Scissor game")                          #Instructions of games
print("\nInstructions")
print("1. Player have to enter following command : ")
print("1 for rock".center(20, '-'))
print("2 for paper".center(20, '-'))
print("3 for scissor".center(20, '-'))
print("4 for exit".center(20, '-'))
print("2. Player will have only 5 chances(lives).")
print("3. Move will be shown in (player's move : computer) format.")
print("4. Please only enter commands like 1,2, 3 or 4.")

steps = ['rock', 'paper', 'scissor']
point = 0
lives = 3

while(lives > 0):     
    print("\n")
    player = int(input('Enter your command : '))
    computer = steps[random.randint(0, len(steps)-1)]
    print("\n")
    lives, point = checkClash(player, computer, lives, point)
     
print("\n")
print("Your final score is : ",point)
```
